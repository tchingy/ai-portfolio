```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

reviews = [
    "The compressor failed completely after just 6 weeks.",
    # NOTE: Model misclassified this review.
    # "Fault" triggered a negative signal even though the meaning is positive.
    # Root cause: model trained on movie reviews, not HVAC service language.
    # Fix: fine-tune on HVAC-labelled data for production use.
    "Engineer arrived early and fixed the fault in under an hour.",
    "Been waiting 3 weeks for a replacement part with no updates.",
    "Very professional team, explained everything clearly before starting.",
    "The system is leaking refrigerant again — same issue as last year.",
    "Annual service was thorough and the report was detailed.",
    "Extremely noisy unit, can hear it from two rooms away.",
    "Excellent response time, had us back up and running same day.",
    "The installation was rushed and now the airflow is uneven.",
    "Best HVAC company I've used — five stars without hesitation."
]

results = classifier(reviews)

for review, result in zip(reviews, results):
    label = result['label']
    score = round(result['score'] * 100, 1)
    print(f"[{label} – {score}%] {review}")

positives = sum(1 for r in results if r['label'] == 'POSITIVE')
negatives = sum(1 for r in results if r['label'] == 'NEGATIVE')
print(f"\nSummary: {positives} positive, {negatives} negative out of {len(reviews)} reviews")
```
